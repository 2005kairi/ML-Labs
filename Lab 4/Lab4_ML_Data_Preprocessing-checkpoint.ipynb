# EDA - Exploratory Data Analysis  
## Online Retail Dataset  
### Dataset Source: Kaggle - Online Retail Dataset  
https://www.kaggle.com/datasets/ulrikthygepedersen/online-retail-dataset

A transaction-level dataset from a UK online retail store containing purchases made from December 2009 to December 2011. It includes product codes, descriptions, quantities, prices, timestamps, customer identifiers, and countries, supporting analysis of sales patterns, customer behavior, product performance, returns, and time-based trends.

### Key Columns  
**InvoiceNo** - Invoice number. Nominal. A 6-digit code uniquely assigned to each transaction; codes starting with "C" indicate cancellations.  
**StockCode** - Product code. Nominal. A 5-digit identifier uniquely assigned to each product.  
**Description** - Product name. Nominal. Text description of the purchased item.  
**Quantity** - Units purchased. Numeric (integer). Number of items per transaction; negative values represent returns.  
**InvoiceDate** - Transaction date. Datetime. Stored as YYYY-MM-DD HH:MM.  
**UnitPrice** - Price per unit in GBP. Numeric (float). Product price per single item.  
**CustomerID** - Customer identifier. Nominal. A 5-digit number uniquely assigned to each customer.  
**Country** - Customer's country. Nominal. Name of the country where each customer resides.

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style="whitegrid")

# Load dataset from the .ipynb_checkpoints folder
df = pd.read_csv(r"C:\Users\Luvje\.ipynb_checkpoints\online_retail.csv", encoding="ISO-8859-1")

# Show first rows
df.head()
```

```python
# viewing the data types of columns
df.dtypes
```

```python
# 2. Data Quality Assessment

# 2.1 View current data types
df.dtypes

# 2.2 Convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# 2.3 Convert CustomerID to string (IDs should not be floats)
df['CustomerID'] = df['CustomerID'].astype('Int64').astype('string')

# 2.4 Ensure Quantity and UnitPrice are numeric
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['UnitPrice'] = pd.to_numeric(df['UnitPrice'], errors='coerce')

# 2.5 Check data types again
df.dtypes
```

```python
# 3. Handling Missing Values

# 3.1 Detect Missing Values
df.isna()          # Shows True/False for each missing value
df.isna().sum()    # Shows total missing values per column
```

```python
df = df.dropna(subset=['CustomerID', 'Description'])
df.isna().sum()
```

**I chose the remove‑records strategy because the missing values are text fields, and mean or median imputation only works for numeric data.**

```python
## 4. Handling Outliers
# We check for extreme values using the IQR method.


# Boxplot for UnitPrice
plt.figure(figsize=(6,4))
sns.boxplot(x=df['UnitPrice'])
plt.title('Boxplot of UnitPrice')
plt.show()

# IQR method
Q1 = df['UnitPrice'].quantile(0.25)
Q3 = df['UnitPrice'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['UnitPrice'] < lower) | (df['UnitPrice'] > upper)]
outliers.head()
```

```python
## 4. Handling Outliers
# We check for extreme values using the IQR method.

import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot for UnitPrice
plt.figure(figsize=(6,4))
sns.boxplot(x=df['UnitPrice'])
plt.title('Boxplot of UnitPrice')
plt.show()

# IQR method
Q1 = df['UnitPrice'].quantile(0.25)
Q3 = df['UnitPrice'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Find outliers
outliers = df[(df['UnitPrice'] < lower) | (df['UnitPrice'] > upper)]

# Show TRUE outliers (sorted)
outliers_sorted = outliers.sort_values('UnitPrice', ascending=False)
outliers_sorted.head()
```

```python
# Keep only values inside the IQR range
df_no_outliers = df[(df['UnitPrice'] >= lower) & (df['UnitPrice'] <= upper)]

print("df with outliers:", df.shape)
print("df without outliers:", df_no_outliers.shape)
```

```python
# Define 5% and 95% caps
lower_cap = df['UnitPrice'].quantile(0.05)
upper_cap = df['UnitPrice'].quantile(0.95)

# Apply capping
df_capped = df.copy()
df_capped['UnitPrice'] = df_capped['UnitPrice'].clip(lower_cap, upper_cap)
```

```python
# Boxplot after capping extreme values
plt.figure(figsize=(6,4))
sns.boxplot(x=df_capped['UnitPrice'])
plt.title('Boxplot of UnitPrice After Capping')
plt.show()
```

```python
# Boxplot after removing outliers (IQR method)
plt.figure(figsize=(6,4))
sns.boxplot(x=df_no_outliers['UnitPrice'])
plt.title('Boxplot of UnitPrice After Removing Outliers (IQR Method)')
plt.show()
```

```python

# STEP 1 — CLEAN BASIC GARBAGE


df_clean = df[df['UnitPrice'] > 0]          # remove zero prices
df_clean = df_clean[df_clean['Quantity'] > 0]  # remove returns
df_clean = df_clean[~df_clean['Description'].str.contains('POSTAGE|Manual', case=False, na=False)]



# STEP 2 — CALCULATE BOXPLOT LIMITS


Q1 = df_clean['UnitPrice'].quantile(0.25)
Q3 = df_clean['UnitPrice'].quantile(0.75)
IQR = Q3 - Q1

lower_whisker = Q1 - 1.5 * IQR
upper_whisker = Q3 + 1.5 * IQR



# STEP 3 — CAPPING AT WHISKERS (THIS REMOVES ALL OUTLIERS)


df_capped = df_clean.copy()
df_capped['UnitPrice'] = df_capped['UnitPrice'].clip(lower_whisker, upper_whisker)



# STEP 4 — BOXPLOT AFTER CAPPING (NO OUTLIERS)


plt.figure(figsize=(6,4))
sns.boxplot(x=df_capped['UnitPrice'])
plt.title('UnitPrice After Capping at Boxplot Whiskers (No Outliers)')
plt.show()
```

**I kept both capping methods to show how different thresholds affect the boxplot: the 5–95% cap reduces extremes, while whisker‑based capping removes all visible outliers.**

```python
## 5.Normalize numerical features using both Min-Max and Z-score.

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_scaled = df[['Quantity', 'UnitPrice']].copy()

df_scaled[['Quantity', 'UnitPrice']] = scaler.fit_transform(df_scaled)

df_scaled.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_standardized = df[['Quantity', 'UnitPrice']].copy()

df_standardized[['Quantity', 'UnitPrice']] = scaler.fit_transform(df_standardized)

df_standardized.head()
```

```python
## 6.Apply PCA only if the numerical features show a correlation.


plt.figure(figsize=(6,4))
sns.heatmap(df_standardized[['Quantity','UnitPrice']].corr(),
            annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Before PCA)")
plt.show()
```

```python
from sklearn.decomposition import PCA

X = df_standardized[['Quantity', 'UnitPrice']]

pca = PCA(n_components=2)
principal_components = pca.fit_transform(X)

print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(principal_components[:,0], principal_components[:,1])
plt.title("PCA Projection")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

**Interpretation of the PCA Projection**

**The PCA scatter plot shows how each transaction is represented using the two principal components. Because the explained variance ratio is almost evenly split (PC1 ≈ 50.06%, PC2 ≈ 49.94%), the points do not stretch strongly in one direction. This indicates that Quantity and UnitPrice do not share meaningful correlation, so PCA does not significantly reduce dimensionality. PCA is applied here mainly for demonstration rather than for compressing the dataset.**
